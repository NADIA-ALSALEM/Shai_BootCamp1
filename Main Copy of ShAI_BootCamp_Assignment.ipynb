```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('Salaries.csv')
df.head()
```

```python
df.columns
```

## Basic Data Exploration

```python
# number of rows and columns in the dataset, data types of each column
# check if there is a missing values in each column
df.info()
num_rows, num_columns = df.shape
print('Number of Rows:' , num_rows)
print('Number of Columns:', num_columns)
missing_values = df.isnull().sum()
print('Missing Values in Each Column:')
print(missing_values)
```

## Descriptive Statistics

```python
# Calculate basic statistics mean, median, mode, minimum, and maximum salary
numeric_columns = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']]
for column in numeric_columns.columns:
    mean_value = numeric_columns[column].mean()
    print(f"Mean {column}: {mean_value}")

    median_value = numeric_columns[column].median()
    print(f"Median {column}: {median_value}")

    mode_value = numeric_columns[column].mode().iloc[0]
    print(f"Mode {column}: {mode_value}")

    min_value = numeric_columns[column].min()
    print(f"Minimum {column}: {min_value}")

    max_value = numeric_columns[column].max()
    print(f"Maximum {column}: {max_value}")

    print()

# determine the range of salaries, and find the standard deviation
salary_column = df['TotalPay']
salary_range = salary_column.max() - salary_column.min()
print('Salary Range :', salary_range)
SalaryStandardDeviation = salary_column.std()
print('Salary Standard Deviation: ',SalaryStandardDeviation)
```

## Data Cleaning

```python
# replace missing value in Otherpay and Overtimepay with mean because it is small
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan , strategy='mean')
imputer.fit(df.iloc[:, 4:6])
df.iloc[: , 4:6] = imputer.transform(df.iloc[: , 4:6])

# remove Notes and Status feature because it is an empty feature and unnecessary in analysis
features_to_remove = ['Notes', 'Status']
df = df.drop(columns=features_to_remove)


# to deal with missing value in "Benefits" and "BasePay" feature i have checked if there is a stronge realtion between them and other feature
# to predict the missing values
correlation = df['Benefits'].corr(df['TotalPay'])
print(f"Correlation between Benefits and TotalPay: {correlation}") # Correlation between Benefits and TotalPay: 0.884 -- > strong relation
correlation = df['BasePay'].corr(df['TotalPay'])
print(f"Correlation between BasePay and TotalPay: {correlation}") # Correlation between BasePay and TotalPay: 0.954 -- > strong relation

# so, i will replace missing value in Benefits and BasePay with mean of non-missing value in TotalPay feature
feature_with_missing = ['Benefits', 'BasePay']
missing_rows = df[feature_with_missing].isnull()
totalpay_for_missing = df.loc[missing_rows.any(axis=1), 'TotalPay']

mean_benefits_for_non_missing_totalpay = df.loc[~missing_rows['Benefits'], 'Benefits'].mean()
mean_basepay_for_non_missing_totalpay = df.loc[~missing_rows['BasePay'], 'BasePay'].mean()

# Fill missing 'Benefits' and 'BasePay' values with the calculated means based on 'TotalPay'
df.loc[missing_rows['Benefits'], 'Benefits'] = mean_benefits_for_non_missing_totalpay
df.loc[missing_rows['BasePay'], 'BasePay'] = mean_basepay_for_non_missing_totalpay


missing_values = df.isnull().sum()
print('Missing Values in Each Column:')
print(missing_values)
```
